````python
import sys
input = sys.stdin.readline
n,m = map(int, input().split())
board = list(input().strip() for _ in range(n))
res = []
for i in range(n-7):
  for j in range(m-7):
    fir_w, fir_b = 0, 0
    for a in range(i, i+8):
      for b in range(j, j+8):
        if (a+b)%2 == 0:
          if board[a][b] != 'W': fir_w += 1
          elif board[a][b] != 'B': fir_b += 1
        else:
          if board[a][b] != 'B': fir_w += 1
          elif board[a][b] != 'W': fir_b += 1
    res.append(min(fir_w,fir_b))
print(min(res))
````

````python
def binSearch(start, end, element, searchList):
  if start > end:
    return 0
  m = (start+end)//2
  if searchList[m] == element:
    return 1
  elif searchList[m] > element:
    return binSearch(start, m-1, element, searchList)
  else:
    return binSearch(m+1, end, element, searchList)

n = int(input())
a = list(map(int, input().split()))
a.sort()
m = int(input())
b = list(map(int, input().split()))

for i in b:
  print(binSearch(0, n-1, i, a))
````

# 4949

image.png

````python
import sys
input = sys.stdin.readline
while(True):
  s = list(input())
  if s[0] == '.':
    break
  res = True
  stack = []
  for i in s:
    if i == '(' or i== '[': stack.append(i)
    elif i == ')':
      if len(stack) > 0 and stack[-1] == '(':
        stack.pop()
      else:
        res = False
    elif i == ']':
      if len(stack) > 0 and stack[-1] == '[':
        stack.pop()
      else:
        res = False
        break
  if len(stack) == 0 and res:
    print("yes")
  else:
    print("no")
````

````python
from bisect import bisect_left, bisect_right

def search(x, card):
    res = bisect_right(card, x) - bisect_left(card, x)
    return res
n = int(input())
card = sorted(list(map(int, input().split())))
m = int(input())
get = list(map(int, input().split()))
for i in get:
    print(search(i, card))
````
