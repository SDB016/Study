```python
import sys
input = sys.stdin.readline
n,m = map(int, input().split())
board = list(input().strip() for _ in range(n))
res = []
for i in range(n-7):
  for j in range(m-7):
    fir_w, fir_b = 0, 0
    for a in range(i, i+8):
      for b in range(j, j+8):
        if (a+b)%2 == 0:
          if board[a][b] != 'W': fir_w += 1
          elif board[a][b] != 'B': fir_b += 1
        else:
          if board[a][b] != 'B': fir_w += 1
          elif board[a][b] != 'W': fir_b += 1
    res.append(min(fir_w,fir_b))
print(min(res))
```

```python
import sys
input = sys.stdin.readline

def check(param):
  for i in range(len(param)//2):
    if param[i] != param[-1*(i+1)]:
      return 'no'
  return 'yes'

while(True):
  s = input().strip()
  if s == '0': break
  print(check(s))
```

# 1654 (어려움)

image.png

```python
import sys
input = sys.stdin.readline
k, n = map(int, input().split())
wires = list(int(input().strip()) for _ in range(k))
start = 1
end = max(wires)
mid = 0
while(start <= end):
  mid = (start+end)//2
  num = 0
  for w in wires:
    num += w//mid
  if num >= n: start = mid + 1
  else: end = mid - 1
print(end)
```

```python
import sys
input = sys.stdin.readline
n = int(input())
des = list(int(input().strip())for _ in range(n))
res = []
k = 0
temp = []

for d in des:
  while(k <= n):
    if len(temp)==0 or temp[-1] != d:
      k += 1
      temp.append(k)
      res.append('+')
    else:
      temp.pop()
      res.append('-')
      break
if len(temp) == 0:
  for i in res:
    print(i)
else:
  print('NO')
```

방법 1: 이분탐색

image.png

```python
def binSearch(start, end, element, searchList):
  if start > end:
    return 0
  m = (start+end)//2
  if searchList[m] == element:
    return 1
  elif searchList[m] > element:
    return binSearch(start, m-1, element, searchList)
  else:
    return binSearch(m+1, end, element, searchList)

n = int(input())
a = list(map(int, input().split()))
a.sort()
m = int(input())
b = list(map(int, input().split()))

for i in b:
  print(binSearch(0, n-1, i, a))
```
