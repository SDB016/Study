```python
import sys
input = sys.stdin.readline
n,m = map(int, input().split())
board = list(input().strip() for _ in range(n))
res = []
for i in range(n-7):
  for j in range(m-7):
    fir_w, fir_b = 0, 0
    for a in range(i, i+8):
      for b in range(j, j+8):
        if (a+b)%2 == 0:
          if board[a][b] != 'W': fir_w += 1
          elif board[a][b] != 'B': fir_b += 1
        else:
          if board[a][b] != 'B': fir_w += 1
          elif board[a][b] != 'W': fir_b += 1
    res.append(min(fir_w,fir_b))
print(min(res))
```

```python
import sys
input = sys.stdin.readline

def check(param):
  for i in range(len(param)//2):
    if param[i] != param[-1*(i+1)]:
      return 'no'
  return 'yes'

while(True):
  s = input().strip()
  if s == '0': break
  print(check(s))
```
