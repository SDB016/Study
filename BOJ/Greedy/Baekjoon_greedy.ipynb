# **17609** (틀린 문제, 어려움)

image.png

```
import sys
def secCheck(word,left,right):
    while(left<right):
        if(word[left]==word[right]):
            left += 1
            right -= 1
        else:
            return False
    return True
def firstCheck(word,left,right):
    while(left<right):
        if(word[left]==word[right]):
            left += 1
            right -= 1
        else:
            check1 = secCheck(word,left+1,right)
            check2 = secCheck(word,left,right-1)
            if(check1 or check2):
                return 1
            else:
                return 2
    return 
n = int(sys.stdin.readline().rstrip("\n"))
for _ in range(n):
    word = sys.stdin.readline().rstrip("\n")
    left, right = 0, len(word)-1
    print(firstCheck(word,left,right))
```

# **1744** (틀린 문제, 다소 어려움)

image.png

```
import sys
minus = []
plus = []
one = 0
def cal(arr):
    res = 0
    for i in range(0,len(arr)-1,2):
        res += arr[i] * arr[i+1]
    if len(arr)%2 != 0:
        res += arr[-1]
    return res
for _ in range(int(sys.stdin.readline().rstrip("\n"))):
    num = int(sys.stdin.readline().rstrip("\n"))
    if num > 1:
        plus.append(num)
    elif num <= 0:
        minus.append(num)
    elif num == 1:
        one += 1
plus.sort(reverse = True)
minus.sort()
print(cal(plus)+cal(minus)+one)
```

# **1764** (틀린 문제)

```
import sys
input = sys.stdin.readline
n, m = map(int,input().rstrip('\n').split())
a = [input().rstrip('\n') for _ in range(n)]
b = [input().rstrip('\n') for _ in range(m)]
dup = sorted(list(set(a) & set(b)))
print(len(dup))
for e in dup:
    print(e)
```

**note:**

1. set에 대해 좀 더 공부하기 (set이외에 자료형과 그에 관한 함수에 대해 좀 더 공부)

2. sort함수와 sorted함수의 차이점

       sort함수:
                  sort함수는 그저 특정 리스트를 정렬해주는 기능을 해주는 함수이다. 반환값이 없는 함수이기 때문에 새로운 객체에 할당하게 될 경우 none이 할당되어 그 객체는 NoneType이 된다.
       
       sorted함수:
                  특정 리스트를 정렬한 후 정렬된 리스트를 반환해주는 함수로 sort함수와 달리 반환값이 존재한다.
    `for i in sorted(lst)`
      처럼 반복문의 list 부분에 활용할 경우 정렬과 동시에 deep copy가 이루어지기 때문에 원본 데이터에 영향을 주지 않는다.
        

# **1080** (틀린 문제)

```
import sys
input = sys.stdin.readline
n, m = map(int,input().split())
cnt = 0
fir = [list(map(int, list(input().rstrip('\n')))) for _ in range(n)]
sec = [list(map(int, list(input().rstrip('\n')))) for _ in range(n)]
def reverse(x,y):
    for i in range(x,x+3):
        for j in range(y,y+3):
            fir[i][j] = 1 - fir[i][j]
def check():
    for i in range(n):
        for j in range(m):
            if fir[i][j] != sec[i][j]:
                return 0
    return 1
res = 0
for i in range(n-2):
    for j in range(m-2):
        if fir[i][j] != sec[i][j]:
            reverse(i,j)
            res += 1
if check():
    print(res)
else:
    print(-1)
```

# **1715**

```
import sys
import heapq
input = sys.stdin.readline
n = int(input())
cards = []
count = 0
for _ in range(n):
    heapq.heappush(cards,int(input()))
while len(cards) > 1:
    tmp= heapq.heappop(cards) + heapq.heappop(cards)
    count += tmp
    heapq.heappush(cards, tmp)
print(count)
```

# **1092** (어려움, 시간초과 주의)

image.png

```
import sys
input = sys.stdin.readline
n = int(input())
limits = list(map(int, input().split()))
m = int(input())
box = list(map(int, input().split()))

time = 0
count = 0
check = [0] * m
pos = [0] * n

limits.sort(reverse= True)
box.sort(reverse= True)

if box[0] > limits[0]:
    print(-1)
else:
    while count < len(box):
        for i in range(n):
            while pos[i] < len(box):
                if not check[pos[i]] and limits[i] >= box[pos[i]]:
                    check[pos[i]] = True
                    pos[i] += 1
                    count += 1
                    break
                pos[i] += 1
        time += 1
    print(time)
            
        
```

# **10775 (어려움, union-find)**

image.png

```
import sys
input = sys.stdin.readline

def findParent(x):
    if x == parent[x]:
        return x
    parent[x] = findParent(parent[x])
    return parent[x]
    
def union(x, y):
    x = findParent(x)
    y = findParent(y)
    if x < y:
        parent[y] = x
    elif x > y:
        parent[x] = y

g = int(input())
p = int(input())
count = 0
plane = list(int(input()) for _ in range(p))
parent = [i for i in range(g+1)]

for i in plane:
    x = findParent(i)
    if x == 0:
        break
    union(x, x-1)
    count += 1
print(count)
```
