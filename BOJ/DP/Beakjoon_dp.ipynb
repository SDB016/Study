```
import sys
input = sys.stdin.readline

n = int(input())
triangles = []
for i in range(n):
    triangles.append(list(map(int, input().split())))
    for j in range(n-len(triangles[i])):
        triangles[i].append(0)
dp = [[0]*501 for _ in range(501)]
dp[0] = triangles[0]
for i in range(1,n):
    for j in range(0,i+1):
        if j==0:
            dp[i][j] = dp[i-1][j]+triangles[i][j]
        elif j==i:
            dp[i][j] = dp[i-1][j-1]+triangles[i][j]
        else:
            dp[i][j] = max(dp[i-1][j-1], dp[i-1][j]) + triangles[i][j]
print(max(dp[n-1]))
```

# **11054**

```
import sys
input = sys.stdin.readline

n = int(int(input()))
a = list(map(int, input().split()))
inc = [1]*n
dec = [1]*n
res = [0]*n
for i in range(n):
    for j in range(i):
        if a[i]>a[j]:
            inc[i] = max(inc[i], inc[j]+1)
for i in range(n-1, -1, -1):
    for j in range(i+1, n):
        if a[i]>a[j]:
            dec[i] = max(dec[i], dec[j]+1)
    res[i] = dec[i]+inc[i]-1
print(max(res))
```

# **9251 (LCS)**

image.png

```
import sys
input = sys.stdin.readline

fir = input().rstrip()
sec = input().rstrip()
fl, sl = len(fir), len(sec)
dp = [[0]*(sl+1) for _ in range(fl+1)]

for i in range(1, fl+1):
    for j in range(1, sl+1):
        if fir[i-1] == sec[j-1]:
            dp[i][j] = dp[i-1][j-1]+1
        else:
            dp[i][j] = max(dp[i-1][j], dp[i][j-1])
print(dp[fl][sl])
```

```
import sys
input = sys.stdin.readline

n = int(int(input()))
a = list(list(map(int, input().split())) for _ in range(n))
dp = [[0]*n for _ in range(n)]
dp[0][0] = 1
for i in range(n):
    for j in range(n):
        if dp[i][j] != 0 and a[i][j] != 0:
            if j+a[i][j] < n:
                dp[i][j+a[i][j]] += dp[i][j]
            if i+a[i][j] < n:
                dp[i+a[i][j]][j] += dp[i][j]
print(dp[n-1][n-1])
```

```
import sys
input = sys.stdin.readline
n,m = map(int, input().split())
arr = list(list(map(int, input().split()))for _ in range(n))
score = [[-1]*m for _ in range(n)]
score[0][0] = arr[0][0]
for i in range(1, n):
    score[i][0] = arr[i][0]+score[i-1][0]
for i in range(1, m):
    score[0][i] = arr[0][i]+score[0][i-1]
    
for i in range(1, n):
    for j in range(1,m):
        score[i][j] = max(score[i-1][j-1], score[i-1][j], score[i][j-1]) + arr[i][j]
print(score[-1][-1])
```

```
n = int(input())
h = []
dp = [[0]*3 for _ in range(n)]
for i in range(n):
    h.append(list(map(int, input().split())))
dp[0] = h[0]
for i in range(1,n):
    for j in range(3):
        dp[i][j] = h[i][j] + min(dp[i-1][(j+1)%3], dp[i-1][(j+2)%3])
print(min(dp[-1]))
```
